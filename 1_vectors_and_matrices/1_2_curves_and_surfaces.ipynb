## Curves in a 2-D space

Curves can be specified in two different ways

### Implicit definition

The solutions to an equation yields a curve in a plane.

For example

$$
y = x^2 - 3
$$

Another example

$$
x^2 + y^2 = 4
$$

### Parametric definition

Specify coordinates as a function of a parameter.

For example

$$
\begin{array}{l}
x(t) &=& t \\
y(t) &=& t^2 - 3
\end{array}
$$

Another example

$$
\begin{array}{l}
x(t) &=& 2 \cos t \\
y(t) &=& 2 \sin t
\end{array}
$$

## Surfaces in a 3-D space

Similarly, a surface in a 3-D space can be defined implicitly and parametrically

### Implicit definition

The solutions of an equation yields a 3-D surface.

For example

$$
z = 3 x^2 + y^2 - 5
$$

Another example

$$
x^2 + y^2 + z^2 = 4
$$

### Parametric definition

Specifying coordinates as a function of **two**

$$
\begin{array}{l}
x(s, t) &=& s \\
y(s, t) &=& t \\
z(s, t) &=& 3 s^2 + t^2 - 5
\end{array}
$$

Another example

$$
\begin{array}{l}
x(s, t) &=& 2 \cos s \sin t \\
y(s, t) &=& 2 \sin s \sin t \\
z(s, t) &=& 2 \cos t
\end{array}
$$


### Surface parametrization

...is a process finding an equivalent parametric definition for an implicit definition of a surface.

For example, find the parametric definition for

$$
2 x^3 - y + 3 z^2 = 5
$$

Let $s, t$ be the parameters. Then a simplest way of parametrization is let

$$
\begin{array}{l}
x(s, t) &=& s \\
z(s, t) &=& t
\end{array}
$$

Then since

$$
y = 2x^3 + 3 z^2 - 5
$$

Then we can write

$$
y(s, t) = 2s^3 + 3 t^2 - 5
$$

Thus the full parametrization is

$$
\begin{array}{l}
x(s, t) &=& s \\
y(s, t) &=& 2s^3 + 3 t^2 - 5 \\
z(s, t) &=& t
\end{array}
$$

### Surface implicitization

...is the inverse process of surface parametrization, which is a process finding an equivalent implicit definition for a parametric definition of a surface.

For example, find the implicit definition for the surface defined as

$$
\begin{array}{l}
x(s, t) &=& s \\
y(s, t) &=& t^2 \\
z(s, t) &=& \sqrt{s^2 - 3 t^2}
\end{array}
$$

We start by substituting $s, t$ by $x, y$ in the equation definition of $z$, then we have

$$
z = \sqrt{x^2 - 3 y}
$$

Thus

$$
z^2 = x^2 - 3 y
$$

...which we can use to infer

$$
3 y = z^2 - x^2
$$

But now, we need to be careful with bounds on the parameters. Recall that $y(s, t) = t^2$ and $z(s, t) = \sqrt{s^2 - 3 t^2}$. Thus $y > 0$ and $z > 0$.

## Quadratic surfaces

A quadratic surface is a generalization of is a generalization of conic sections (ellipses, parabolas, and hyperbolas).

### Spheres

A sphere with radius $r$ and center $(x_0, z_0, y_0)$ is defined implicit as

$$
(x - x_0)^2 + (y - y_0)^2 + (z - z_0)^2 = r^2
$$

Or equivalently

$$
\frac{(x - x_0)^2}{r^2} + \frac{(y - y_0)^2}{r^2} + \frac{(z - z_0)^2}{r^2} = 1
$$

The second form is useful for being modified easily to yield the other quadratic surfaces.

### Ellipsoids

The axis-aligned ellipsoid with semi-axis "radii" $a, b, c > 0$ and center $(x_0, y_0, z_0)$ is defined as

$$
\frac{(x - x_0)^2}{a^2} + \frac{(y - y_0)^2}{b^2} + \frac{(z - z_0)^2}{c^2} = 1
$$

If $a = b < c$ then this is a prolate spheroid.

If $a = b > c$ then this is an oblate spheroid.

If $a = b = c$ then this is a sphere.

### Hyperboloids

#### One-sheeted hyperboloid

When one of the signs in the equation of the ellipsoid is turned negative, one has an one-sheeted hyperboloid.

$$
\frac{(x - x_0)^2}{a^2} + \frac{(y - y_0)^2}{b^2} - \frac{(z - z_0)^2}{c^2} = 1
$$

The negative term indicates which axis the hyperboloid is open along.

#### Two-sheeted hyperboloid

When two of the signs in the equation of the ellipsoid is turned negative, one has an two-sheeted hyperboloid.

$$
\frac{(x - x_0)^2}{a^2} - \frac{(y - y_0)^2}{b^2} - \frac{(z - z_0)^2}{c^2} = 1
$$

In this case the positive term indicates which axis the hyperboloid is open along.

### Paraboloids

#### Elliptic paraboloids

An elliptic paraboloid opens along one axis, and is an ellipse in cross-section in that axis.

$$
z - z_0 = \frac{(x - x_0)^2}{a^2} + \frac{(y - y_0)^2}{b^2}
$$

In this case the paraboloid opens along the $z$-axis, towards its positive direction.

#### Hyperbolic paraboloids

An hyperbolic paraboloid is parabolic along one axis, and is a hyperbola in cross-section in that axis.

$$
z - z_0 = \frac{(x - x_0)^2}{a^2} - \frac{(y - y_0)^2}{b^2}
$$

In this case the paraboloid is parabolic along the $z$-axis, and is a hyperbola in the cross-section of the $z$-axis.

### Cones

A cone opens along one surface, and is an ellipse in cross-section to that axis.

$$
\frac{(z - z_0)^2}{c^2} = \frac{(x - x_0)^2}{a^2} + \frac{(y - y_0)^2}{b^2}
$$

In this case the cone opens along the $z$-axis, and is an ellipse in cross-section to the $z$-axis.

### Cylinders

#### Elliptic cylinder

$$
\frac{x^2}{a^2} + \frac{y^2}{b^2} = 1
$$

#### Hyperbolic cylinder

$$
\frac{x^2}{a^2} - \frac{y^2}{b^2} = 1
$$

#### Parabolic cylinder

$$
x^2 + 2 a y = 0
$$
